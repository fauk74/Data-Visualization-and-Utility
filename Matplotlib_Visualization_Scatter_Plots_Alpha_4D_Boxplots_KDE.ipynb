# Bars and Histograms

# Axes and Legend

```python
df2['c'].plot(figsize=(20,3),ls=':',c='red', lw=5)
```

```python
title="My Plot Title"
xlabel='My X Data'
ylabel='My y Data'
ax=df2['c'].plot(figsize=(20,3),ls=':',c='red', lw=5)
```

```python
ax=df2['c'].plot(figsize=(20,3),ls=':',c='red', lw=5)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

```python
ax=df2['c'].plot(figsize=(20,3),ls=':',c='red', lw=5)
ax.set(xlabel=xlabel,ylabel=ylabel)
ax.legend(loc=2, bbox_to_anchor=(1.0,1.0)) # legend outside the box
```
